cylinders ,model year,origin are catorical variable

```
dummies_cylinder=pd.get_dummies(df.cylinders,drop_first=True,prefix='cylinder')
```

```
dummies_origin=pd.get_dummies(df.origin,prefix='origin')
```

```
from sklearn.preprocessing import LabelEncoder
```

```
labelEncoding=LabelEncoder()
```

```
df['horsepower']=labelEncoding.fit_transform(df['horsepower'])
```

```
df['Model_year']=labelEncoding.fit_transform(df['model year'])
df
```

```
df=pd.concat([df,dummies_cylinder,dummies_origin],axis=1)
```

Drop the catorical variable

```
remove_list=['cylinders','model year','origin','car name']
```

```
df.drop(remove_list,axis=1,inplace=True)
```

# **Train Test and Split**

```
X=df.drop('mpg',axis=1)
```

```
y=df.mpg
```

```
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```
from sklearn.linear_model import LinearRegression
```

```
linear_model=LinearRegression()
```

```
linear_model.fit(X_train,y_train)
```

```
prediction=linear_model.predict(X_test)
```

```
from sklearn.metrics import mean_squared_error
```

```
mean_squared_error(y_test,prediction)
```

```
from sklearn.metrics import r2_score
```

```
r2_score(y_test,prediction)
```
